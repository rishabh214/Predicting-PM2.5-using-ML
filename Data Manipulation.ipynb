```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt 

df=pd.read_csv('data1.csv')
#df['pm2.5'][30]# 30 is the index here.mentioned on dataframe
#df['temp'].iloc[0] this coomand will work for normal indexes
```

```python
df.head()
```

```python
from datetime import datetime
## changeing str to datetime
c=0

for  i in df['start']:

    datetime_object = datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
    df['start'][c]=datetime_object
    c+=1
c=0

for  i in df['end']:

    datetime_object = datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
    df['end'][c]=datetime_object
    c+=1
```

## missing value treatment


```python
df.isnull().sum()
```

```python
df.dropna(thresh=11,inplace=True)
df['nox'].fillna(df['nox'].mean(),inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
file=pd.pivot_table(df,['pm2.5','pm10','co','o3','so2'],'temp')
```

```python
file
```

```python
a=df['pm2.5']
a.fillna(0,inplace=True)
c=df['pm10']
c.fillna(0,inplace=True)
b=df['co']
b.fillna(0,inplace=True)
for i in range(0,7449):
   
    if (a.iloc[i]==0):
        a.iloc[i]=a.iloc[i-1]    #filling previoyus data
    
    if (c.iloc[i]==0):
        c.iloc[i]=c.iloc[i-1]
   
    if (b.iloc[i]==0):
        t=df['temp'].iloc[i]
        b.iloc[i]=file['co'][t]  # filling according temp
```

```python
df['o3'].fillna(df['o3'].mean(),inplace=True)
df['so2'].fillna(df['so2'].mean(),inplace= True)
```

# normalization


```python
x=df[['pm10', 'temp', 'bp', 'rh', 'ws', 'co', 'o3','nox', 'so2']]
x=x.apply(lambda x: (x-x.min(axis=0) ) / (x.max(axis=0)-x.min(axis=0)))
```

```python
x=x.round(3)
```

```python
from datetime import datetime
## changeing str to datetime
c=0

for  i in df['start']:

    datetime_object = datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
    df['start'][c]=datetime_object
    c+=1
c=0

for  i in df['end']:

    datetime_object = datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
    df['end'][c]=datetime_object
    c+=1
```

```python
for i in range(8431):
    df['date'].iloc[i]=str(df['start'].iloc[i].day )+'/'+str(df['start'].iloc[i].month)
```

## missing value treatment


```python
df.isnull().sum()
```

```python
df.dropna(thresh=11,inplace=False)
df['so2'].fillna(df['so2'].mean(),inplace=True)
df.shape
```

```python
df.dropna(how='any',inplace=True)
```

```python
a=df['pm2.5']
a.fillna(0,inplace=True)
c=df['pm10']
c.fillna(0,inplace=True)
b=df['co']
b.fillna(0,inplace=True)
d=df['o3']
d.fillna(0,inplace=True)
e=df['nox']
e.fillna(0,inplace=True)
for i in range(0,7449):
   
    if (a.iloc[i]==0):
        a.iloc[i]=a.iloc[i-1]    #filling previoyus data
    
    if (c.iloc[i]==0):
        c.iloc[i]=c.iloc[i-1]
    
    if (b.iloc[i]==0):
        b.iloc[i]=b.iloc[i-1]
    
    if (d.iloc[i]==0):
        d.iloc[i]=d.iloc[i-1]
        
    if (e.iloc[i]==0):
        e.iloc[i]=e.iloc[i-1]
   
#     if (b.iloc[i]==0):
#         t=df['temp'].iloc[i]
#         b.iloc[i]=file['co'][t]  # filling according temp
# df['o3'].fillna(df['o3'].mean(),inplace=True)
# df['so2'].fillna(df['so2'].mean(),inplace= True)
```

# normalization


```python
x=df[['pm10', 'temp', 'bp', 'rh', 'ws', 'co', 'o3','nox', 'so2']]
x=x.apply(lambda x: (x-x.min(axis=0) ) / (x.max(axis=0)-x.min(axis=0)))
```
